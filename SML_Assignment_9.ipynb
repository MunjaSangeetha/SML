```python
import pandas as pd
data=pd.read_csv('/content/breast_cancer_survival.csv')
data
```

```python
data=data.fillna(0)
```

```python
replace=data.replace({'FEMALE':1,'MALE':0})
replace
```

```python
replace1 = replace.replace({'Alive': 1, 'Dead': 0,'I':1,'II':2,"III":3})
replace1

```

```python
replace2=replace1.replace({'Positive':1,'Negative':0})
replace2
```

```python
replace3 = replace2.replace({('Infiltrating Ductal Carcinoma'): 1, ('Infiltrating Lobular Carcinoma'): 2,('Mucinous Carcinoma'):3, 'Other': 1, 'Lumpectomy': 1, 'Modified Radical Mastectomy': 1, 'Simple Mastectomy': 1})
replace3
```

```python
features=replace3.drop('Patient_Status',axis=1)
f=features
f=f.drop('Date_of_Surgery',axis=1)
f=f.drop('Date_of_Last_Visit',axis=1)
print(f)
```

```python
target=replace3['Patient_Status']
target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(f,target,test_size=0.2,random_state=20)
```

LOGISTIC REGREESION


```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import seaborn as sns
yp=model.predict(x_test)
print(confusion_matrix(y_test,yp))
```

```python
print(accuracy_score(y_test,yp))
print(classification_report(y_test,yp))
print(confusion_matrix(y_test,yp))
```
